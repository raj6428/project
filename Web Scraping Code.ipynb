```python
#Import required modules
import time
from selenium import webdriver
from selenium.common.exceptions import NoSuchElementException,ElementNotInteractableException
from selenium.webdriver.support.select import Select
import pandas as pd
import re
import warnings
from selenium.webdriver.common.action_chains import ActionChains
warnings.filterwarnings('ignore')
```

```python
#Initialising Chrome web driver
driver=webdriver.Chrome('I:\chromedriver.exe')

#Website
url='https://www.yatra.com/'

#getting the website
driver.get(url)
driver.maximize_window()
time.sleep(5)

#Search button 
search= driver.find_element_by_xpath('//*[@id="BE_flight_flsearch_btn"]').click()
time.sleep(5)

#Details scrapped
airline=[]
departure_time=[]
departure_city=[]
reaching_time=[]
destination_city=[]
price=[]
journey_time=[]
no_of_stops=[]
others=[]
date_of_journey=[]

cities=['hyderabad','chennai','bangalore','goa','delhi','trichy']

for city in cities:
    dept=driver.find_element_by_id('origin_0')
    dept.clear()
    time.sleep(2)
    dept.send_keys(city)
    dept.click()
    time.sleep(2)
    
    search=driver.find_element_by_xpath('//button[@class="fs-14 btn-submit cursor-pointer bold"]').click()
    time.sleep(5)

    for j in range(1,8):
        # date change  
        try:
            date = driver.find_element_by_xpath(f'//*[@id="Flight-APP"]/section/section[2]/section[1]/div[1]/div/div[1]/div/ul/li[{j}]/div/p[2]').click()
            time.sleep(10)
        except ElementNotInteractableException:
            pass

        #Get scroll height after first time page load
        last_height = driver.execute_script("return document.body.scrollHeight")
        while True:
            # Scroll down to bottom
            driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
            # Wait to load page
            time.sleep(3)
            # Calculate new scroll height and compare with last scroll height
            new_height = driver.execute_script("return document.body.scrollHeight")
            if new_height == last_height:
                break
            last_height = new_height

        #Flight names
        airlines=driver.find_elements_by_xpath('//div[@class="fs-13 airline-name no-pad col-8"]/span')
        for i in airlines:
            airline.append(i.text)
        print('Length of airlines:',len(airline))

        #departure time
        departure = driver.find_elements_by_xpath('//div[@class="i-b pr"]')
        for dept in departure:
            departure_time.append(dept.text)
        print('Length of departure time:',len(departure_time))

        #departure city
        city_from = driver.find_elements_by_xpath('//div[@class="i-b col-4 no-wrap text-right dtime col-3"]/p')
        for i in city_from:
            departure_city.append(i.text)
        print('Length of departure city:',len(departure_city))

        #destination city
        dest_city = driver.find_elements_by_xpath('//div[@class="i-b pdd-0 text-left atime col-5"]/p[2]')
        for i in dest_city:
            destination_city.append(i.text)
        print('Length of destination city:',len(destination_city))

        #Destination landing time
        reach_time=driver.find_elements_by_xpath('//p[@class="bold fs-15 mb-2 pr time"]')
        for i in reach_time:
            reaching_time.append(i.text)
        print('Length of reaching time:',len(reaching_time))

        #Ticket fare
        ticket_price= driver.find_elements_by_xpath('//div[@class="i-b tipsy fare-summary-tooltip fs-18"]')
        for prices in ticket_price:
            price.append(prices.text)
        print('Length of price:',len(price))

        #journey_time
        journey = driver.find_elements_by_xpath('//p[@class="fs-12 bold du mb-2"]')
        for i in journey:
            journey_time.append(i.text)
        print('Length of journey time:',len(journey_time))

        #no_of_stops
        no_stops= driver.find_elements_by_xpath('//div[@class=" font-lightgrey fs-10 tipsy i-b fs-10"]')
        for i in no_stops:
            no_of_stops.append(i.text)
        print('Length of number of stops:',len(no_of_stops))

        #other details
        try:
            other_details=driver.find_elements_by_xpath('//div[@class="flight-footer  ml-24 mr-24"]')
            for i in other_details:
                others.append(i.text)
        except NoSuchElementException:
            others.append('No details')
        print('Length of other details:',len(others))

        #scrolling to top
        driver.execute_script("window.scrollTo(document.body.scrollHeight,-10)")
        time.sleep(5)

        #Date of journey
        dojs=driver.find_element_by_xpath('//div[@class="day-li text-center cursor-pointer pr active font-primary-color"]/p')
        for i in range(len(other_details)):
            date_of_journey.append(dojs.text)
        print(len(date_of_journey))
        time.sleep(3)


    
        
```

```python
flights_data =pd.DataFrame({})
flights_data['airline']=airline
flights_data['date_of_journey']=date_of_journey
flights_data['departure_city']=departure_city
flights_data['departure_time']=departure_time
flights_data['reaching_time']=reaching_time
flights_data['stops']=no_of_stops
flights_data['destination_city']=destination_city
flights_data['journey_time']=journey_time
flights_data['others']=others
flights_data['ticket fare']=price
flights_data
```

```python
flights_data.to_csv('airlines.csv')
```
